```python
from sklearn.datasets import load_iris
```

```python
from sklearn.decomposition import KernelPCA
```

```python
import pandas
```

```python
df = load_iris(as_frame = True)['data']
df['type'] = load_iris(as_frame = True)['target']

df
```

```python
df.describe()
```

```python
X = load_iris(as_frame = True)['data']
```

```python
kpca = KernelPCA(n_components=2, kernel='rbf')
```

```python
#KernelPCA? - utilizando os parametros podemos alterar os resultados obtidos
```

```python
kpca.fit(X)
```

```python
result_kpca = pandas.DataFrame(kpca.transform(X))
result_kpca.columns = ["kpca1", "kpca2"]
result_kpca['type'] = df['type']
```

```python
import plotly.express as px
```

```python
fig = px.scatter(result_kpca, x='kpca1', y='kpca2', color='type', title='kpca com k=2, kernel = rbf')

fig.show()
```

```python
# alterando o valor de gamma:

kpca = KernelPCA(n_components=2, kernel='rbf', gamma=0.1)
```

```python
kpca.fit(X)
```

```python
result_kpca = pandas.DataFrame(kpca.transform(X))
result_kpca.columns = ["kpca1", "kpca2"]
result_kpca['type'] = df['type']
```

```python
fig = px.scatter(result_kpca, x='kpca1', y='kpca2', color='type', title='kpca com k=2, kernel = rbf')

fig.show()
```
