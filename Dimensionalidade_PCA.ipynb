```python
# Importando dados
from sklearn.datasets import load_iris
```

```python
# PCA - é uma tecnica estatistica utilizada para reduzir a dimensionalidade de um conjunto de dados, preservando ao máximo a sua variabilidade.
# Esta técnica reduz variaveis linearmente correlacionadas. Exemplo extremo é se tivermos duas variaves iguais, uma em centimetros e outra em metros
# por serem diretamente correlacionadas, podem ser reduzidas em apenas uma.
```

```python
# Importanto PCA

from sklearn.decomposition import PCA
```

```python
import pandas
```

```python
df = load_iris(as_frame = True)['data']
df['type'] = load_iris(as_frame = True)['target']

df
```

```python
# Por termos 4 variaveis, nao conseguimos ter uma visualização gráfica dos dados, por isso, precisamos fazer uma dimensionalidade para obter esta visualização

# Para isto, vamos escolher o k=2
```

```python
df.describe()
```

```python
# aqui podemos observar que existe uma variancia muito grande entre os dados, o que dificulta a utilização do PCA. A componente que possui maior variancia
# provavelmente irá ser dominante quando for posto para identificar as componentes principais
```

```python
# caracteristicas da pétala:

X = load_iris(as_frame = True)['data']
```

```python
# instanciando o pca com k=2 "algoritmo vazio"

pca = PCA(n_components=2)
```

```python
# treinando o algotimo:

pca.fit(X)
```

```python
# imprimindo a variancia explicada:

print(pca.explained_variance_ratio_)
```

```python
# ou seja, aqui vemos que a primeira componente principal explicou 92,42% de toda a variabilidade dos dados e a segunda 5,3%
# somando temos que elas explicaram 97,7% da variabilidade de todos os dados.
```

```python
pca.transform(X)
```

```python
# para cada observação, estes são os componentes principais
```

```python
result_pca = pandas.DataFrame(pca.transform(X))
result_pca.columns = ["pca1", "pca2"]
result_pca["type"] = df["type"] # coluna para difenciar os tipos de iris
```

```python
import plotly.express as px
```

```python
fig = px.scatter(result_pca, x = 'pca1', y = 'pca2', color = "type", title = 'PCA com k=2')

fig.show()
```

```python
# avaliando o gráfico, vemos que o tipo 0 esta bem distante do tipo 1 e do tipo 2, por isso será o mais fácil de ser identificado.

# para variancias muito diferentes, sabemos que o pca nao funciona tão bem, por isso iremos padronizar os dados para realizar uma nova analise
```

```python
# vamos deixar os dados media 0 e variancia 1 e depois testar novamente
# isso é feito pegando a variavel, diminuindo a media dela e dividindo pelo desvio padrao. Isso para todas as variaveis

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
pca_scaled = PCA(n_components = 2)
```

```python
pandas.DataFrame(X_scaled).describe()
```

```python
# agora, com o desvio padrao igual a 1 para todas as variaveis, podemos dizer que todas elas possuem a mesma variancia. Isso ajuda ao PCA a nao dar
# mais importancia a uma informação do que outra
```

```python
pca_scaled.fit(X_scaled)
```

```python
print(pca_scaled.explained_variance_ratio_)
```

```python
# agora vemos que a variancia explicada ficou mais equilibrada. totalizando 95,81%
```

```python
result_pca_scaled = pandas.DataFrame(pca_scaled.transform(X_scaled))
result_pca_scaled.columns = ["pca1", "pca2"]
result_pca_scaled["type"] = df["type"] # coluna para difenciar os tipos de iris
```

```python
result_pca_scaled.head()
```

```python
fig = px.scatter(result_pca_scaled, x='pca1', y='pca2', color = 'type', title = "PCA com k=2, dados na mesma escala")

fig.show()
```

```python
# mesmo com o pca alterado, notamos uma dificuldade em destinguir os tipos 1 e 2
```
